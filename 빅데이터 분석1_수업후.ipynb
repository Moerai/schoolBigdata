#### 영화 정보를 data/movie.txt에서 읽어오도록 수정

```python
# 영화추천 함수 수정
영화파일경로 = 'data/movie.data'
영화 = {}
for 줄 in open(영화파일경로, encoding='utf-8-sig'):
    영화제목, 영화장르 = 줄.split('|')
    영화장르 = 영화장르.strip().split(',')
    영화[영화제목] = 영화장르
print(영화)
```

```python
def 영화추천(장르):
    추천영화 = []
    영화파일경로 = 'data/movie.data'
    영화 = {}
    for 줄 in open(영화파일경로, encoding='utf-8-sig'):
        영화제목, 영화장르 = 줄.split('|')
        영화장르 = 영화장르.strip().split(',')
        영화[영화제목] = 영화장르
    for 영화제목, 영화장르 in 영화.items():
        if 장르 in 영화장르:
            추천영화.append(영화제목)
    return 추천영화
```

## 2) IRIS 데이터 분석

```python
# 실습. 각 종 별로, 4개의 특징의 평균값을 구하라.
print(data[0])
```

```python
분석결과 = {}
for 라벨, 수치 in data:
    if 라벨 in 분석결과.keys():
        횟수 = 분석결과[라벨][4] #가장 마지막에 해당 라벨의 데이터 수를 저장.
        집계목록 = []
        for i in range(4):
            집계값 = (분석결과[라벨][i] * 횟수 + 수치[i])/(횟수 + 1)
            집계값 = float('{0:.3f}'.format(집계값))
            #포맷 = '{0:.2f}'
            #집계값 = 포맷.format(집계값)
            집계목록.append(집계값)
        집계목록.append(횟수+1)
        분석결과[라벨] = 집계목록
    else:
        분석결과[라벨] = 수치 + [1] 
print(분석결과)        
        
```

## 3) 파일 출력

```python
파일 = open('data/iris_analysis1.data', 'w')
for 라벨, 수치 in 분석결과.items():
    line = ','.join([str(n) for n in 수치] + [라벨])
    line += '\n'
    파일.write(line)
파일.close() #close() 해야지만, 파일이 생성됨. 
```
